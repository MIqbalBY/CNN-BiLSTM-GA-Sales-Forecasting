## 🌱 2. Global Seed Lock — Reproducibility

> **Mengapa seed harus dikunci?**  
> Deep learning melibatkan banyak sumber keacakan: inisialisasi bobot, dropout, shuffle mini-batch, dan operasi CUDA.  
> Tanpa seed yang konsisten, hasil eksperimen tidak dapat direproduksi — bahkan di mesin yang sama.
>
> Pipeline ini mengunci **5 lapisan** sumber keacakan:
>
> | Layer | Target |
> |-------|--------|
> | `PYTHONHASHSEED` | Hash built-in Python |
> | `random.seed()` | Modul random standar |
> | `numpy.random.seed()` | Operasi NumPy |
> | `tf.random.set_seed()` | Operasi TensorFlow/Keras |
> | `tf.config.experimental.enable_op_determinism()` | CUDA op level |
>
> **Nilai seed:** `GLOBAL_SEED = 42` (digunakan konsisten di seluruh pipeline)

````python
# ── Konfigurasi GLOBAL_SEED ───────────────────────────────────
GLOBAL_SEED = 42

# Layer 1: Python hash seed
os.environ['PYTHONHASHSEED'] = str(GLOBAL_SEED)

# Layer 2: Python built-in random
random.seed(GLOBAL_SEED)

# Layer 3: NumPy
np.random.seed(GLOBAL_SEED)

# Layer 4: TensorFlow / Keras
tf.random.set_seed(GLOBAL_SEED)

# Layer 5: TF deterministic ops (GPU level)
try:
    tf.config.experimental.enable_op_determinism()
    determinism_status = "✅ ENABLED"
except Exception:
    # Tersedia sejak TF 2.9; older versions: partially deterministic
    determinism_status = "⚠️  Partial (TF < 2.9, upgrade recommended)"

# ── Konfirmasi ────────────────────────────────────────────────
print("=" * 60)
print("  GLOBAL SEED LOCK — REPRODUCIBILITY TERJAMIN")
print("=" * 60)
print(f"  GLOBAL_SEED              : {GLOBAL_SEED}")
print(f"  PYTHONHASHSEED           : {os.environ['PYTHONHASHSEED']}  ✅")
print(f"  random.seed()            : ✅ SET")
print(f"  numpy.random.seed()      : ✅ SET")
print(f"  tf.random.set_seed()     : ✅ SET")
print(f"  TF Op Determinism        : {determinism_status}")
print("=" * 60)

# Verifikasi konsistensi numerik
np.random.seed(GLOBAL_SEED)  # reset untuk verifikasi
verify_sample = np.random.rand(3)
print(f"  Verifikasi NumPy random  : {verify_sample}")
print(f"  (nilai di atas harus konsisten di setiap run)")
print("=" * 60)
````

## 💻 3. Deteksi Hardware

> Spesifikasi hardware dicatat sebagai bagian dari metadata eksperimen.  
> Jika GPU tersedia, **memory growth** diaktifkan untuk mencegah alokasi memori penuh sekaligus.

````python
# ── CPU Info ──────────────────────────────────────────────────
cpu_count   = multiprocessing.cpu_count()
cpu_info    = platform.processor() or "N/A"

# ── GPU Info ──────────────────────────────────────────────────
gpus = tf.config.list_physical_devices('GPU')
gpu_details = []

if gpus:
    for gpu in gpus:
        try:
            details = tf.config.experimental.get_device_details(gpu)
            gpu_details.append(f"  → {gpu.name} — {details.get('device_name', 'Unknown')}")
        except Exception:
            gpu_details.append(f"  → {gpu.name}")
    # Enable memory growth (hindari OOM)
    for gpu in gpus:
        try:
            tf.config.experimental.set_memory_growth(gpu, True)
        except RuntimeError:
            pass
    gpu_mode   = f"{len(gpus)} GPU — Memory Growth ENABLED"
    gpu_note   = ""
else:
    gpu_details = ["  GPU Mode            : CPU-only mode (menggunakan TF CPU backend)"]
    gpu_mode    = "CPU-only mode (menggunakan TF CPU backend)"
    gpu_note    = (
        "\n  ℹ️  Catatan: Model akan berjalan di CPU.\n"
        "     Training GA mungkin memerlukan waktu lebih lama.\n"
        "     Pertimbangkan Google Colab (T4 GPU) untuk training GA."
    )

# ── Print ─────────────────────────────────────────────────────
print("=" * 60)
print("  DETEKSI HARDWARE")
print("=" * 60)
print(f"  CPU Cores (Logical) : {cpu_count}")
print(f"  CPU Processor       : {cpu_info}")
print(f"  RAM Total (approx)  : Cek via psutil (opsional)")
print(f"  GPU Terdeteksi      : {len(gpus)} unit")
for gd in gpu_details:
    print(gd)
print(f"  TF Built with CUDA  : {tf.test.is_built_with_cuda()}")
print(f"  GPU Available       : {len(gpus) > 0}")
if gpu_note:
    print("-" * 60 + gpu_note)
print("=" * 60)

# ── Simpan ke dict global ─────────────────────────────────────
HARDWARE_INFO = {
    "cpu_count"       : cpu_count,
    "cpu_info"        : cpu_info,
    "gpu_count"       : len(gpus),
    "gpu_names"       : [g.name for g in gpus],
    "tf_cuda_support" : bool(tf.test.is_built_with_cuda()),
    "gpu_available"   : len(gpus) > 0,
    "compute_mode"    : "GPU" if gpus else "CPU"
}
````
